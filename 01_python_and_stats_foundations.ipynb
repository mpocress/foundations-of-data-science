## Part 1: Python Essentials

Before diving into data science libraries, let's make sure we're solid on core Python. These skills — reading files, parsing text, organizing data — are things you'll use constantly.

### 1.1 Working with Strings and Parsing

Weather data often arrives as messy text. Below is a raw string representing a week of temperature readings. Each line contains a date and a high/low temperature in Fahrenheit.

**Exercise:** Parse this string into a list of dictionaries, where each dictionary has keys `'date'`, `'high'`, and `'low'`.

```python
raw_weather = """2024-01-15,34,21
2024-01-16,29,18
2024-01-17,38,25
2024-01-18,42,30
2024-01-19,37,28
2024-01-20,31,19
2024-01-21,28,14"""

# YOUR CODE HERE
# Parse raw_weather into a list of dicts called `weather_records`
# Each dict should look like: {'date': '2024-01-15', 'high': 34, 'low': 21}

weather_records = []

# --- Solution ---
for line in raw_weather.strip().split('\n'):
    parts = line.split(',')
    weather_records.append({
        'date': parts[0],
        'high': int(parts[1]),
        'low': int(parts[2])
    })

print(weather_records[:3])
```

### 1.2 List Comprehensions and Transformations

List comprehensions are a Pythonic way to transform and filter data in a single expression.

**Exercise:** Using `weather_records` from above:
1. Create a list of daily temperature **ranges** (high - low) using a list comprehension.
2. Create a list of dates where the high was above freezing (> 32°F).

```python
# YOUR CODE HERE

# 1. Daily temperature ranges
temp_ranges = [r['high'] - r['low'] for r in weather_records]
print("Temperature ranges:", temp_ranges)

# 2. Dates above freezing
above_freezing = [r['date'] for r in weather_records if r['high'] > 32]
print("Days above freezing:", above_freezing)
```

### 1.3 Functions

Writing reusable functions is essential. Below, write two utility functions we'll use later.

**Exercise:**
1. Write `fahrenheit_to_celsius(f)` that converts a Fahrenheit value to Celsius.
   - Formula: C = (F - 32) × 5/9
2. Write `describe_temps(records)` that takes a list of weather record dicts and returns a dictionary with the `'mean_high'`, `'mean_low'`, and `'max_range'`.

```python
# YOUR CODE HERE

def fahrenheit_to_celsius(f):
    """Convert Fahrenheit to Celsius."""
    return (f - 32) * 5 / 9

def describe_temps(records):
    """Return summary statistics from a list of weather record dicts."""
    highs = [r['high'] for r in records]
    lows = [r['low'] for r in records]
    ranges = [r['high'] - r['low'] for r in records]
    return {
        'mean_high': sum(highs) / len(highs),
        'mean_low': sum(lows) / len(lows),
        'max_range': max(ranges)
    }

# Test
print(f"32°F = {fahrenheit_to_celsius(32):.1f}°C")
print(f"212°F = {fahrenheit_to_celsius(212):.1f}°C")
print(describe_temps(weather_records))
```

### 1.4 Classes

Classes let you bundle data and behavior together. This is useful when you want to represent a concept (like a weather station) with both state and operations.

**Exercise:** Complete the `WeatherStation` class below.

```python
class WeatherStation:
    """Represents a weather station that collects daily temperature readings."""
    
    def __init__(self, name, location):
        self.name = name
        self.location = location
        self.readings = []  # list of (date, high, low) tuples
    
    def add_reading(self, date, high, low):
        """Add a daily temperature reading."""
        self.readings.append((date, high, low))
    
    def average_high(self):
        """Return the average high temperature across all readings."""
        if not self.readings:
            return None
        return sum(r[1] for r in self.readings) / len(self.readings)
    
    def coldest_day(self):
        """Return the date with the lowest low temperature."""
        if not self.readings:
            return None
        coldest = min(self.readings, key=lambda r: r[2])
        return coldest[0]
    
    def __repr__(self):
        return f"WeatherStation('{self.name}', readings={len(self.readings)})"


# Test it
station = WeatherStation("Boston Logan", "Boston, MA")
for r in weather_records:
    station.add_reading(r['date'], r['high'], r['low'])

print(station)
print(f"Average high: {station.average_high():.1f}°F")
print(f"Coldest day: {station.coldest_day()}")
```

## Part 2: NumPy Fundamentals

NumPy is the backbone of numerical computing in Python. It provides fast, vectorized operations on arrays — essential when you're working with thousands (or millions) of data points.

### 2.1 Array Creation and Basic Operations

Let's simulate 365 days of temperature data and practice array operations.

**Exercise:** Study the code below, then answer the questions that follow.

```python
np.random.seed(42)

# Simulate daily high temps: seasonal pattern + random noise
days = np.arange(365)
seasonal = 55 + 25 * np.sin(2 * np.pi * (days - 80) / 365)  # peaks in summer
noise = np.random.normal(0, 5, size=365)
daily_highs = seasonal + noise

print(f"Shape: {daily_highs.shape}")
print(f"Min: {daily_highs.min():.1f}°F")
print(f"Max: {daily_highs.max():.1f}°F")
print(f"Mean: {daily_highs.mean():.1f}°F")
print(f"Std: {daily_highs.std():.1f}°F")
```

### 2.2 Vectorized Operations

One of NumPy's key advantages is that you can apply operations to entire arrays at once — no loops needed.

**Exercise:**
1. Convert `daily_highs` from Fahrenheit to Celsius (store as `daily_highs_c`).
2. Count how many days had highs above 80°F.
3. Find the indices of the 5 hottest days.

```python
# YOUR CODE HERE

# 1. Convert to Celsius
daily_highs_c = (daily_highs - 32) * 5 / 9
print(f"Mean high in Celsius: {daily_highs_c.mean():.1f}°C")

# 2. Days above 80°F
hot_days = np.sum(daily_highs > 80)
print(f"Days above 80°F: {hot_days}")

# 3. Indices of 5 hottest days
hottest_indices = np.argsort(daily_highs)[-5:]
print(f"5 hottest days (day of year): {hottest_indices}")
print(f"Their temperatures: {daily_highs[hottest_indices]}")
```

### 2.3 Linear Algebra Basics

Many machine learning algorithms rely on linear algebra under the hood. Let's practice a few key operations.

**Exercise:**
1. Create two vectors `a = [1, 2, 3]` and `b = [4, 5, 6]`. Compute their **dot product**.
2. Create a 3×3 matrix `M` and compute its **inverse** and **determinant**.
3. Solve the linear system `Mx = v` where `v = [1, 2, 3]`.

```python
# YOUR CODE HERE

# 1. Dot product
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(f"Dot product: {np.dot(a, b)}")

# 2. Matrix inverse and determinant
M = np.array([[2, 1, 0],
              [1, 3, 1],
              [0, 1, 2]])

M_inv = np.linalg.inv(M)
det_M = np.linalg.det(M)
print(f"\nDeterminant of M: {det_M:.2f}")
print(f"Inverse of M:\n{M_inv}")

# 3. Solve Mx = v
v = np.array([1, 2, 3])
x = np.linalg.solve(M, v)
print(f"\nSolution x: {x}")
print(f"Verification M @ x: {M @ x}")  # should equal v
```

## Part 3: Pandas — Tabular Data Manipulation

Pandas is the go-to library for working with structured data in Python. Let's build a DataFrame from our simulated weather data and practice the operations you'll use daily.

### 3.1 Creating and Exploring a DataFrame

**Exercise:** Run the cell below to create a weather DataFrame, then answer the exploration questions.

```python
np.random.seed(42)

dates = pd.date_range('2024-01-01', periods=365, freq='D')
days_num = np.arange(365)

# Simulate temperatures with seasonal patterns
high_temps = 55 + 25 * np.sin(2 * np.pi * (days_num - 80) / 365) + np.random.normal(0, 5, 365)
low_temps = high_temps - np.random.uniform(8, 20, 365)
precipitation = np.random.exponential(0.15, 365)  # inches
humidity = 50 + 20 * np.sin(2 * np.pi * (days_num - 120) / 365) + np.random.normal(0, 8, 365)

weather_df = pd.DataFrame({
    'date': dates,
    'high_temp_f': np.round(high_temps, 1),
    'low_temp_f': np.round(low_temps, 1),
    'precip_inches': np.round(precipitation, 2),
    'humidity_pct': np.clip(np.round(humidity, 1), 10, 100)
})

# Add month and season columns
weather_df['month'] = weather_df['date'].dt.month
weather_df['season'] = weather_df['month'].map(
    {12: 'Winter', 1: 'Winter', 2: 'Winter',
     3: 'Spring', 4: 'Spring', 5: 'Spring',
     6: 'Summer', 7: 'Summer', 8: 'Summer',
     9: 'Fall', 10: 'Fall', 11: 'Fall'}
)

weather_df.head(10)
```

```python
# Explore the DataFrame
print(f"Shape: {weather_df.shape}")
print(f"\nColumn types:\n{weather_df.dtypes}")
print(f"\nBasic statistics:")
weather_df.describe()
```

### 3.2 Filtering and Selection

**Exercise:**
1. Select all rows where the high temperature exceeded 85°F.
2. Find the 10 days with the most precipitation.
3. Select all Summer days where humidity was below 40%.

```python
# YOUR CODE HERE

# 1. Days above 85°F
hot_days_df = weather_df[weather_df['high_temp_f'] > 85]
print(f"Days above 85°F: {len(hot_days_df)}")
print(hot_days_df[['date', 'high_temp_f']].head())

print("\n---")

# 2. Top 10 precipitation days
top_rain = weather_df.nlargest(10, 'precip_inches')[['date', 'precip_inches']]
print("\nTop 10 wettest days:")
print(top_rain)

print("\n---")

# 3. Dry summer days
dry_summer = weather_df[(weather_df['season'] == 'Summer') & (weather_df['humidity_pct'] < 40)]
print(f"\nDry summer days (humidity < 40%): {len(dry_summer)}")
```

### 3.3 Grouping and Aggregation

Grouping lets you compute statistics for subsets of your data — one of the most powerful Pandas operations.

**Exercise:**
1. Compute the **mean high temperature** for each season.
2. Compute the **total precipitation** for each month.
3. Find the **day with the highest temperature range** (high - low) in each season.

```python
# YOUR CODE HERE

# 1. Mean high by season
print("Mean high temperature by season:")
print(weather_df.groupby('season')['high_temp_f'].mean().round(1))

print("\n---")

# 2. Total precipitation by month
print("\nTotal precipitation by month:")
monthly_precip = weather_df.groupby('month')['precip_inches'].sum().round(2)
print(monthly_precip)

print("\n---")

# 3. Largest daily temperature range per season
weather_df['temp_range'] = weather_df['high_temp_f'] - weather_df['low_temp_f']
idx = weather_df.groupby('season')['temp_range'].idxmax()
print("\nDay with largest temp range per season:")
print(weather_df.loc[idx, ['date', 'season', 'high_temp_f', 'low_temp_f', 'temp_range']])
```

## Part 4: Statistics Foundations

Data science is built on statistics. Here we'll practice the core concepts: distributions, descriptive statistics, and basic inference.

### 4.1 Descriptive Statistics

**Exercise:** Compute the following for the `high_temp_f` column:
1. Mean, median, and standard deviation
2. The 25th and 75th percentiles (and the IQR)
3. Skewness — is the distribution symmetric?

```python
# YOUR CODE HERE

temps = weather_df['high_temp_f']

# 1. Central tendency and spread
print(f"Mean:   {temps.mean():.2f}°F")
print(f"Median: {temps.median():.2f}°F")
print(f"Std:    {temps.std():.2f}°F")

# 2. Percentiles and IQR
q25 = temps.quantile(0.25)
q75 = temps.quantile(0.75)
iqr = q75 - q25
print(f"\n25th percentile: {q25:.2f}°F")
print(f"75th percentile: {q75:.2f}°F")
print(f"IQR: {iqr:.2f}°F")

# 3. Skewness
print(f"\nSkewness: {temps.skew():.3f}")
print("(Values near 0 suggest symmetry)")
```

### 4.2 Probability Distributions

Understanding distributions is key to statistical modeling. Let's visualize the distribution of temperatures and fit a normal distribution to them.

**Exercise:** Plot a histogram of `high_temp_f` and overlay a fitted normal distribution curve.

```python
# YOUR CODE HERE

fig, ax = plt.subplots()

# Histogram
ax.hist(temps, bins=30, density=True, alpha=0.7, color='steelblue', edgecolor='white')

# Fit and overlay normal distribution
mu, sigma = stats.norm.fit(temps)
x = np.linspace(temps.min(), temps.max(), 100)
ax.plot(x, stats.norm.pdf(x, mu, sigma), 'r-', linewidth=2, 
        label=f'Normal fit ($\mu$={mu:.1f}, $\sigma$={sigma:.1f})')

ax.set_xlabel('High Temperature (°F)')
ax.set_ylabel('Density')
ax.set_title('Distribution of Daily High Temperatures')
ax.legend()
plt.tight_layout()
plt.show()
```

### 4.3 Correlation

Correlation measures the linear relationship between two variables. Values range from -1 (perfect negative) to +1 (perfect positive).

**Exercise:** 
1. Compute the correlation between `high_temp_f` and `humidity_pct`.
2. Create a scatter plot of high temperature vs. humidity.
3. Interpret: does the relationship make physical sense?

```python
# YOUR CODE HERE

# 1. Correlation
corr = weather_df['high_temp_f'].corr(weather_df['humidity_pct'])
print(f"Correlation (temp vs humidity): {corr:.3f}")

# 2. Scatter plot
fig, ax = plt.subplots()
ax.scatter(weather_df['high_temp_f'], weather_df['humidity_pct'], 
           alpha=0.4, s=15, color='teal')
ax.set_xlabel('High Temperature (°F)')
ax.set_ylabel('Humidity (%)')
ax.set_title(f'Temperature vs. Humidity (r = {corr:.3f})')
plt.tight_layout()
plt.show()
```

### 4.4 Hypothesis Testing

Hypothesis tests help us decide whether an observed pattern is statistically meaningful or could have occurred by chance.

**Exercise:** Test whether the mean high temperature in Summer is significantly different from the mean high temperature in Winter. Use a two-sample t-test.

- **Null hypothesis (H₀):** The mean summer high equals the mean winter high.
- **Alternative hypothesis (H₁):** They are different.

```python
# YOUR CODE HERE

summer_temps = weather_df[weather_df['season'] == 'Summer']['high_temp_f']
winter_temps = weather_df[weather_df['season'] == 'Winter']['high_temp_f']

print(f"Summer mean: {summer_temps.mean():.1f}°F (n={len(summer_temps)})")
print(f"Winter mean: {winter_temps.mean():.1f}°F (n={len(winter_temps)})")

t_stat, p_value = stats.ttest_ind(summer_temps, winter_temps)
print(f"\nt-statistic: {t_stat:.2f}")
print(f"p-value: {p_value:.2e}")

alpha = 0.05
if p_value < alpha:
    print(f"\nResult: Reject H₀ at α={alpha}. The difference is statistically significant.")
else:
    print(f"\nResult: Fail to reject H₀ at α={alpha}. No significant difference found.")
```

## Part 5: Visualization

Good visualizations are a core data science skill. Let's create a few common plot types using our weather data.

### 5.1 Line Plot — Temperature Over Time

**Exercise:** Plot the daily high and low temperatures over the full year.

```python
fig, ax = plt.subplots(figsize=(12, 5))

ax.plot(weather_df['date'], weather_df['high_temp_f'], color='tomato', 
        alpha=0.7, linewidth=0.8, label='High')
ax.plot(weather_df['date'], weather_df['low_temp_f'], color='steelblue', 
        alpha=0.7, linewidth=0.8, label='Low')
ax.fill_between(weather_df['date'], weather_df['low_temp_f'], weather_df['high_temp_f'],
                alpha=0.15, color='gray')

ax.set_xlabel('Date')
ax.set_ylabel('Temperature (°F)')
ax.set_title('Daily High and Low Temperatures (2024)')
ax.legend()
plt.tight_layout()
plt.show()
```

### 5.2 Box Plot — Temperature by Season

**Exercise:** Create a box plot showing the distribution of high temperatures across seasons.

```python
fig, ax = plt.subplots()

season_order = ['Winter', 'Spring', 'Summer', 'Fall']
data_by_season = [weather_df[weather_df['season'] == s]['high_temp_f'] for s in season_order]

bp = ax.boxplot(data_by_season, labels=season_order, patch_artist=True)
colors = ['#4a90d9', '#7bc47f', '#e8564a', '#e8a94a']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_ylabel('High Temperature (°F)')
ax.set_title('High Temperature Distribution by Season')
plt.tight_layout()
plt.show()
```

### 5.3 Bar Chart — Monthly Precipitation

**Exercise:** Create a bar chart showing total precipitation per month.

```python
fig, ax = plt.subplots()

monthly_precip = weather_df.groupby('month')['precip_inches'].sum()
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

ax.bar(range(1, 13), monthly_precip.values, color='steelblue', alpha=0.8, edgecolor='white')
ax.set_xticks(range(1, 13))
ax.set_xticklabels(month_labels)
ax.set_xlabel('Month')
ax.set_ylabel('Total Precipitation (inches)')
ax.set_title('Monthly Precipitation (2024)')
plt.tight_layout()
plt.show()
```
